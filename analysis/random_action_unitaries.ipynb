```python
import numpy as np
from relaqs import RESULTS_DIR
import qutip
```

## 100 episodes

```python
#data = np.load(RESULTS_DIR + "random_actions_unitaries_noiseless.npy")
data = np.load(RESULTS_DIR + "random_actions_unitaries_noiseless-100-episodes.npy")
unitaries = [x.reshape((2, 2)) for x in data]   
final_states = [np.matmul(unitary, np.array([1, 0])) for unitary in unitaries]
random_q_objs = [qutip.Qobj(state) for state in final_states]

b = qutip.Bloch()
b.add_states(random_q_objs)
b.show()
```
